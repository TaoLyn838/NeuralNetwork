```python
import numpy as np
import pandas as pd
from simplePerceptron import SimplePerceptron
import matplotlib.pyplot as plt
```

```python
data = [1, 0, 1, 0, 1]
epochs = 2000

perceptron = SimplePerceptron(np.array(data))
predict_list = perceptron.train(epochs)
```

```python
plt.plot(predict_list)
plt.xlabel("Epoch")
plt.ylabel("loss value")
plt.title("Simple perceptron")
plt.show()
```
